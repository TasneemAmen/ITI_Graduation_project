```python
import pandas as pd
import numpy as np

# =========================
# 1. Load Data
# =========================

file_path = "New Cairo KPIs.xlsx"

df = pd.read_excel(file_path)
print("done")
```

```python
# =========================
# 2. Define Columns
# =========================
print("no of cells: ", df["Cell Name"].nunique())
print("no of eNodeBs: ", df["eNodeB Name"].nunique())
print("no of LocalCellsID: ", df["LocalCell Id"].nunique())
print("no of Clusters: ", df["Cluster"].nunique())
print("date range: ", df["Date"].min(), " to ", df["Date"].max())
print("total records: ", len(df))
print("data Dimentions: ", df.shape)
date_col = "Date"
cell_col = "Cell Name"
site_col = "eNodeB Name"
local_cell_col = "LocalCell Id"
cluster_col = "Cluster"

target_kpi = "(HU) DL Traffic Volume (GBytes)"

needed_cols = [
    date_col,
    site_col,
    cell_col,
    local_cell_col,
    cluster_col,
    target_kpi
]
```

```python
# =========================
# 3. Clean Data
# =========================

df.loc[:, date_col] = pd.to_datetime(df[date_col], errors="coerce")
# instead of the line below to suppress a warining for pandas 3.0!
# df[date_col] = pd.to_datetime(df[date_col], errors="coerce")

df_dl = df[needed_cols].copy()

df_dl = df_dl.dropna(subset=[date_col, target_kpi])

df_dl.loc[:, target_kpi] = pd.to_numeric(df_dl[target_kpi], errors="coerce")
#same here, in order to work for pandas  3.0
#df_dl[target_kpi] = pd.to_numeric(df_dl[target_kpi], errors="coerce")

df_dl = df_dl.dropna(subset=[target_kpi])

df_dl = df_dl[df_dl[target_kpi] >= 0]
```

```python
# =========================
# 4. Define Time Periods
# =========================

last_date = df_dl[date_col].max()

# Number of days to compare including the day in question
num_days = 3

# Last num_days in the dataset
recent_start_date = last_date - pd.Timedelta(days=num_days - 1)
recent_end_date = last_date

# Same num_days last week
baseline_start_date = recent_start_date - pd.Timedelta(days=7)
baseline_end_date = recent_end_date - pd.Timedelta(days=7)

print("Last date in data:", last_date.date())
print("Recent period:", recent_start_date.date(), "to", recent_end_date.date())
print("Baseline period:", baseline_start_date.date(), "to", baseline_end_date.date())
```

```python
# =========================
# 5. Filter Recent and Baseline
# =========================

recent_df = df_dl[
    (df_dl[date_col] >= recent_start_date) &
    (df_dl[date_col] <= recent_end_date)
].copy()

baseline_df = df_dl[
    (df_dl[date_col] >= baseline_start_date) &
    (df_dl[date_col] <= baseline_end_date)
].copy()
```

```python
# =========================
# 6. Aggregate Per Cell
# =========================

cell_id_cols = [
    site_col,
    cell_col,
    local_cell_col
]

recent_agg = recent_df.groupby(cell_id_cols).agg(
    recent_avg_dl_traffic=(target_kpi, "mean"),
    recent_total_dl_traffic=(target_kpi, "sum"),
    recent_days_count=(date_col, "nunique")
).reset_index()

baseline_agg = baseline_df.groupby(cell_id_cols).agg(
    baseline_avg_dl_traffic=(target_kpi, "mean"),
    baseline_total_dl_traffic=(target_kpi, "sum"),
    baseline_days_count=(date_col, "nunique")
).reset_index()
```

```python
# =========================
# 7. Merge Both Periods
# =========================

dl_comparison = recent_agg.merge(
    baseline_agg,
    on=cell_id_cols,
    how="inner"
)

# Remove cells with zero baseline traffic
dl_comparison = dl_comparison[
    dl_comparison["baseline_avg_dl_traffic"] > 0
].copy()

```

```python
# =========================
# 8. Calculate Degradation Ratio
# =========================

dl_comparison["dl_traffic_degradation_ratio_%"] = (
    (dl_comparison["baseline_avg_dl_traffic"] - dl_comparison["recent_avg_dl_traffic"])
    / dl_comparison["baseline_avg_dl_traffic"]
) * 100
```

```python
# =========================
# 9. Classify Degraded Cells
# =========================

degradation_threshold = 20 # %

dl_comparison["dl_traffic_status"] = np.where(
    dl_comparison["dl_traffic_degradation_ratio_%"] >= degradation_threshold,
    "Degraded",
    "Normal"
)
```

```python
# =========================
# 10. Extract Degraded Cells
# =========================

dl_degraded_cells = dl_comparison[
    dl_comparison["dl_traffic_status"] == "Degraded"
].copy()

dl_degraded_cells["degraded_kpi"] = target_kpi



dl_degraded_cells = dl_degraded_cells.sort_values(
    by="dl_traffic_degradation_ratio_%",
    ascending=False
)
```

```python
# ============================================================
# ALGORITHM 01 - DEFINE RELATED COUNTERS FOR DL TRAFFIC
# ============================================================
# This section contains the counters/KPIs that may explain
# why DL Traffic Volume is degraded.
#
# You can edit:
# - threshold
# - reason
# - recommended_parameter
#
# bad_direction:
# "low"  means the problem happens when recent value is lower
# "high" means the problem happens when recent value is higher
# ============================================================

dl_related_rules = [
    {
        "feature": "(HU) Cell DL Average Throughput (Mbps)",
        "bad_direction": "low",
        "threshold": 20,
        "category": "DL Throughput Degradation",
        "reason": "Cell DL throughput decreased compared with the same days last week.",
        "recommended_parameter": "Check DL scheduler, bandwidth, CA activation, load balancing, and congestion."
    },
    {
        "feature": "(HU) User DL Average Throughput (Mbps)",
        "bad_direction": "low",
        "threshold": 20,
        "category": "User Experience Degradation",
        "reason": "User DL throughput decreased compared with the same days last week.",
        "recommended_parameter": "Check radio quality, PRB load, scheduler behavior, and user distribution."
    },
    {
        "feature": "(HU) DL PRB Utilization(%)",
        "bad_direction": "high",
        "threshold": 20,
        "category": "Capacity / Congestion",
        "reason": "DL PRB utilization increased, which may indicate congestion.",
        "recommended_parameter": "Check load balancing parameters, CA usage, bandwidth, and capacity expansion."
    },
    {
        "feature": "DL Average CQI",
        "bad_direction": "low",
        "threshold": 15,
        "category": "Radio Quality Issue",
        "reason": "DL CQI decreased, which may indicate poor radio quality.",
        "recommended_parameter": "Check interference, PCI conflict, antenna tilt, azimuth, and coverage."
    },
    {
        "feature": "(HU) DL IBLER(%)",
        "bad_direction": "high",
        "threshold": 20,
        "category": "DL Radio Quality Issue",
        "reason": "DL IBLER increased, which may indicate retransmission and bad DL quality.",
        "recommended_parameter": "Check interference, CQI, MCS, antenna tilt, and DL power."
    },
    {
        "feature": "DL RBLER",
        "bad_direction": "high",
        "threshold": 20,
        "category": "DL Radio Failure",
        "reason": "DL RBLER increased, which may indicate DL radio block errors.",
        "recommended_parameter": "Check DL interference, poor coverage, CQI, MCS, and radio conditions."
    },
    {
        "feature": "(HU) PDSCH MCS",
        "bad_direction": "low",
        "threshold": 15,
        "category": "Poor Modulation Efficiency",
        "reason": "PDSCH MCS decreased, which may indicate poor radio conditions.",
        "recommended_parameter": "Check CQI, SINR, interference, antenna direction, and coverage."
    },
    {
        "feature": "Availability",
        "bad_direction": "low",
        "threshold": 1,
        "category": "Availability Issue",
        "reason": "Cell availability decreased.",
        "recommended_parameter": "Check cell outage, alarms, S1 issue, manual outage, and system unavailability."
    },
    {
        "feature": "(HU) Cell Unavail Time (s)",
        "bad_direction": "high",
        "threshold": 20,
        "category": "Cell Unavailability",
        "reason": "Cell unavailable time increased.",
        "recommended_parameter": "Check site outage, alarms, transmission issue, power issue, and S1 failure."
    },
    {
        "feature": "L.Traffic.ActiveUser.Dl.Avg",
        "bad_direction": "low",
        "threshold": 20,
        "category": "Traffic Demand Drop",
        "reason": "DL active users decreased, so DL traffic drop may be caused by lower demand.",
        "recommended_parameter": "Check if this is normal traffic behavior before applying RF optimization."
    },
    {
        "feature": "MAC CA Traffic Ratio",
        "bad_direction": "low",
        "threshold": 20,
        "category": "Carrier Aggregation Issue",
        "reason": "CA traffic ratio decreased, which may reduce DL traffic capacity.",
        "recommended_parameter": "Check CA activation, SCell availability, CA parameters, and supported bands."
    },
    {
        "feature": "DL Traffic QCI-9",
        "bad_direction": "low",
        "threshold": 20,
        "category": "Default Bearer Traffic Drop",
        "reason": "QCI-9 DL traffic decreased, which may explain total DL traffic degradation.",
        "recommended_parameter": "Check data service traffic, user demand, packet data service, and internet traffic trend."
    },
    {
        "feature": "DL_CCE_AllocFail (%)",
        "bad_direction": "high",
        "threshold": 20,
        "category": "Control Channel Congestion",
        "reason": "DL CCE allocation failure increased, which may affect scheduling and throughput.",
        "recommended_parameter": "Check PDCCH/CCE utilization, control channel capacity, and scheduler configuration."
    }
]
```

```python
# ============================================================
# ALGORITHM 02 - GET AVAILABLE RELATED FEATURES
# ============================================================
# This avoids errors if any feature does not exist in your Excel file.
# Only existing columns will be used.
# ============================================================

available_related_features = [
    rule["feature"] for rule in dl_related_rules
    if rule["feature"] in df.columns
]

print("Number of available related features:", len(available_related_features))

for feature in available_related_features:
    print("-", feature)
```

```python
# ============================================================
# ALGORITHM 03 - PREPARE RELATED COUNTERS DATA
# ============================================================
# This section creates a new dataframe containing:
# - cell identity columns
# - date column
# - related counters/features
# ============================================================

analysis_cols = cell_id_cols + [date_col] + available_related_features

df_reason = df[analysis_cols].copy()

# Convert date column to datetime and remove time
df_reason[date_col] = pd.to_datetime(
    df_reason[date_col],
    errors="coerce"
).dt.normalize()

# Convert all related counters to numeric values
# This handles values like:
# "95%"
# "1,200"
# " 30.5 "
for col in available_related_features:
    df_reason[col] = (
        df_reason[col]
        .astype(str)
        .str.replace("%", "", regex=False)
        .str.replace(",", "", regex=False)
        .str.strip()
    )

    df_reason[col] = pd.to_numeric(df_reason[col], errors="coerce")

print("Related counter data prepared successfully")
print("Rows:", df_reason.shape[0])
print("Columns:", df_reason.shape[1])
```

```python
# ============================================================
# ALGORITHM 04 - FILTER RELATED COUNTERS BY TIME PERIOD
# ============================================================
# We use the same recent and baseline dates that you already defined.
#
# recent_start_date
# recent_end_date
# baseline_start_date
# baseline_end_date
# ============================================================

recent_reason_df = df_reason[
    (df_reason[date_col] >= recent_start_date) &
    (df_reason[date_col] <= recent_end_date)
].copy()

baseline_reason_df = df_reason[
    (df_reason[date_col] >= baseline_start_date) &
    (df_reason[date_col] <= baseline_end_date)
].copy()

print("Recent reason rows:", recent_reason_df.shape[0])
print("Baseline reason rows:", baseline_reason_df.shape[0])
```

```python
# ============================================================
# ALGORITHM 05 - AGGREGATE RELATED COUNTERS PER CELL
# ============================================================
# For each degraded cell, we calculate the average value of
# each related counter in:
# - recent period
# - baseline period
# ============================================================

recent_reason_agg = (
    recent_reason_df
    .groupby(cell_id_cols)[available_related_features]
    .mean()
    .reset_index()
)

baseline_reason_agg = (
    baseline_reason_df
    .groupby(cell_id_cols)[available_related_features]
    .mean()
    .reset_index()
)

# Rename columns to separate recent values from baseline values
recent_reason_agg = recent_reason_agg.rename(
    columns={col: f"recent_{col}" for col in available_related_features}
)

baseline_reason_agg = baseline_reason_agg.rename(
    columns={col: f"baseline_{col}" for col in available_related_features}
)

print("Recent related counters aggregated:", recent_reason_agg.shape)
print("Baseline related counters aggregated:", baseline_reason_agg.shape)
```

```python
# ============================================================
# ALGORITHM 06 - MERGE RELATED COUNTERS WITH DEGRADED CELLS
# ============================================================
# This adds recent and baseline counter values to each degraded cell.
# ============================================================

dl_degraded_cells_with_causes = dl_degraded_cells.merge(
    recent_reason_agg,
    on=cell_id_cols,
    how="left"
)

dl_degraded_cells_with_causes = dl_degraded_cells_with_causes.merge(
    baseline_reason_agg,
    on=cell_id_cols,
    how="left"
)

print("Degraded cells merged with related counters")
print("Rows:", dl_degraded_cells_with_causes.shape[0])
print("Columns:", dl_degraded_cells_with_causes.shape[1])
```

```python
# ============================================================
# ALGORITHM 07 - FUNCTION TO CALCULATE COUNTER DEGRADATION
# ============================================================
# This function calculates how much a counter became worse.
#
# If bad_direction = "low":
# Example:
# Baseline CQI = 10
# Recent CQI = 7
# Degradation = 30%
#
# If bad_direction = "high":
# Example:
# Baseline PRB = 60
# Recent PRB = 90
# Degradation = 50%
# ============================================================

def calculate_counter_change(recent_value, baseline_value, bad_direction):
    if pd.isna(recent_value) or pd.isna(baseline_value):
        return np.nan

    if baseline_value == 0:
        return np.nan

    if bad_direction == "low":
        change = ((baseline_value - recent_value) / baseline_value) * 100

    elif bad_direction == "high":
        change = ((recent_value - baseline_value) / baseline_value) * 100

    else:
        change = np.nan

    return change
```

```python
# ============================================================
# ALGORITHM 08 - FUNCTION TO FIND MULTIPLE DEGRADATION CAUSES
# ============================================================
# This function checks all related counters for each degraded cell.
#
# It returns:
# - main cause
# - main cause value
# - number of detected causes
# - all detected causes
# - recommended actions
# ============================================================

def find_dl_degradation_causes(row, rules):
    detected_causes = []

    for rule in rules:
        feature = rule["feature"]

        recent_col = f"recent_{feature}"
        baseline_col = f"baseline_{feature}"

        if recent_col not in row.index or baseline_col not in row.index:
            continue

        recent_value = row[recent_col]
        baseline_value = row[baseline_col]

        change_pct = calculate_counter_change(
            recent_value,
            baseline_value,
            rule["bad_direction"]
        )

        if pd.isna(change_pct):
            continue

        if change_pct >= rule["threshold"]:
            detected_causes.append({
                "feature": feature,
                "recent_value": recent_value,
                "baseline_value": baseline_value,
                "change_pct": change_pct,
                "category": rule["category"],
                "reason": rule["reason"],
                "recommended_parameter": rule["recommended_parameter"]
            })

    # If no related counter passed its threshold
    if len(detected_causes) == 0:
        return pd.Series({
            "main_cause_counter_or_kpi": "No strong related counter detected",
            "main_cause_recent_value": np.nan,
            "main_cause_baseline_value": np.nan,
            "main_cause_change_%": np.nan,
            "main_root_cause_category": "Unknown",
            "main_degradation_reason": "DL traffic decreased, but no related counter passed the degradation threshold.",
            "main_recommended_parameter_or_action": "Check raw counters, alarms, availability, and traffic behavior manually.",
            "number_of_detected_causes": 0,
            "multi_cause_flag": "No",
            "all_detected_causes": "None",
            "all_cause_categories": "Unknown",
            "all_recommended_actions": "Manual investigation needed"
        })

    # Sort causes by highest degradation percentage
    detected_causes = sorted(
        detected_causes,
        key=lambda x: x["change_pct"],
        reverse=True
    )

    main_cause = detected_causes[0]

    all_causes_text = " | ".join([
        f"{cause['feature']}: recent={cause['recent_value']:.2f}, "
        f"baseline={cause['baseline_value']:.2f}, "
        f"change={cause['change_pct']:.2f}%"
        for cause in detected_causes
    ])

    all_categories_text = " | ".join([
        cause["category"] for cause in detected_causes
    ])

    all_actions_text = " | ".join([
        cause["recommended_parameter"] for cause in detected_causes
    ])

    return pd.Series({
        "main_cause_counter_or_kpi": main_cause["feature"],
        "main_cause_recent_value": main_cause["recent_value"],
        "main_cause_baseline_value": main_cause["baseline_value"],
        "main_cause_change_%": main_cause["change_pct"],
        "main_root_cause_category": main_cause["category"],
        "main_degradation_reason": main_cause["reason"],
        "main_recommended_parameter_or_action": main_cause["recommended_parameter"],
        "number_of_detected_causes": len(detected_causes),
        "multi_cause_flag": "Yes" if len(detected_causes) > 1 else "No",
        "all_detected_causes": all_causes_text,
        "all_cause_categories": all_categories_text,
        "all_recommended_actions": all_actions_text
    })
```

```python
# ============================================================
# ALGORITHM 09 - APPLY MULTI-CAUSE ANALYSIS
# ============================================================
# This applies the smart algorithm on every degraded cell.
# ============================================================

cause_results = dl_degraded_cells_with_causes.apply(
    lambda row: find_dl_degradation_causes(row, dl_related_rules),
    axis=1
)

dl_degraded_cells_with_causes = pd.concat(
    [dl_degraded_cells_with_causes, cause_results],
    axis=1
)

print("Multi-cause root analysis completed")
```

```python
# ============================================================
# ALGORITHM 10 - RECOMMEND PARAMETERS / ACTIONS BASED ON CAUSE
# ============================================================
# This section adds optimization recommendations based on:
# - main_root_cause_category
# - main_cause_counter_or_kpi
#
# It creates new columns:
# - recommended_action_type
# - recommended_parameters_to_check
# - detailed_recommended_action
# - action_priority
# ============================================================


def recommend_action_based_on_cause(row):
    category = row.get("main_root_cause_category", "")
    cause = row.get("main_cause_counter_or_kpi", "")
    degradation_ratio = row.get("dl_traffic_degradation_ratio_%", 0)

    # -------------------------------
    # Set priority based on degradation
    # -------------------------------
    if degradation_ratio >= 70:
        priority = "P1 - Critical"
    elif degradation_ratio >= 50:
        priority = "P2 - High"
    elif degradation_ratio >= 30:
        priority = "P3 - Medium"
    else:
        priority = "P4 - Low"

    # -------------------------------
    # Capacity / Congestion
    # -------------------------------
    if category == "Capacity / Congestion":
        return pd.Series({
            "recommended_action_type": "Capacity Optimization",
            "recommended_parameters_to_check": (
                "DL PRB Utilization, Load Balancing, Carrier Aggregation, "
                "Bandwidth, Scheduler, Traffic Distribution"
            ),
            "detailed_recommended_action": (
                "DL traffic dropped while DL PRB utilization increased. "
                "Check if the cell is congested. Review load balancing settings, "
                "carrier aggregation activation, bandwidth configuration, and scheduler behavior. "
                "If congestion is persistent, consider capacity expansion or traffic offloading."
            ),
            "action_priority": priority
        })

    # -------------------------------
    # DL Throughput Degradation
    # -------------------------------
    elif category == "DL Throughput Degradation":
        return pd.Series({
            "recommended_action_type": "Throughput Optimization",
            "recommended_parameters_to_check": (
                "DL Scheduler, DL PRB, CQI, MCS, BLER, CA, Bandwidth"
            ),
            "detailed_recommended_action": (
                "Cell DL throughput decreased together with DL traffic degradation. "
                "Check DL scheduler behavior, PRB utilization, CQI, MCS, BLER, and CA usage. "
                "If radio quality is normal but throughput is low, investigate congestion or scheduler limitation."
            ),
            "action_priority": priority
        })

    # -------------------------------
    # User Experience Degradation
    # -------------------------------
    elif category == "User Experience Degradation":
        return pd.Series({
            "recommended_action_type": "User Throughput Optimization",
            "recommended_parameters_to_check": (
                "User DL Throughput, Active Users, CQI, PRB Utilization, Scheduler, CA"
            ),
            "detailed_recommended_action": (
                "User DL throughput decreased, which means user experience is affected. "
                "Check whether the issue is caused by high number of users, bad radio quality, "
                "high PRB load, or low CA usage. Review scheduler and traffic distribution."
            ),
            "action_priority": priority
        })

    # -------------------------------
    # Radio Quality Issue
    # -------------------------------
    elif category == "Radio Quality Issue":
        return pd.Series({
            "recommended_action_type": "Radio Quality Optimization",
            "recommended_parameters_to_check": (
                "CQI, SINR, RSRP, RSRQ, PCI, Antenna Tilt, Azimuth, DL Power, Neighbor Interference"
            ),
            "detailed_recommended_action": (
                "DL CQI decreased, which may indicate poor radio quality or interference. "
                "Check PCI collision/confusion, overshooting cells, antenna tilt, azimuth, "
                "coverage level, and interference from neighboring cells."
            ),
            "action_priority": priority
        })

    # -------------------------------
    # DL Radio Quality Issue
    # -------------------------------
    elif category == "DL Radio Quality Issue":
        return pd.Series({
            "recommended_action_type": "BLER / Interference Optimization",
            "recommended_parameters_to_check": (
                "DL IBLER, DL RBLER, CQI, PDSCH MCS, DL Power, Interference, Antenna Tilt"
            ),
            "detailed_recommended_action": (
                "DL IBLER increased, which means more downlink retransmissions. "
                "Check interference, CQI degradation, PDSCH MCS drop, antenna tilt, "
                "DL power settings, and coverage quality."
            ),
            "action_priority": priority
        })

    # -------------------------------
    # DL Radio Failure
    # -------------------------------
    elif category == "DL Radio Failure":
        return pd.Series({
            "recommended_action_type": "Radio Failure Investigation",
            "recommended_parameters_to_check": (
                "DL RBLER, CQI, MCS, RSRP, RSRQ, Interference, Coverage"
            ),
            "detailed_recommended_action": (
                "DL RBLER increased, which may indicate downlink radio block errors. "
                "Check poor coverage, interference, CQI/MCS degradation, and radio conditions."
            ),
            "action_priority": priority
        })

    # -------------------------------
    # Poor Modulation Efficiency
    # -------------------------------
    elif category == "Poor Modulation Efficiency":
        return pd.Series({
            "recommended_action_type": "MCS / Radio Efficiency Optimization",
            "recommended_parameters_to_check": (
                "PDSCH MCS, CQI, SINR, BLER, Antenna Tilt, Azimuth, Interference"
            ),
            "detailed_recommended_action": (
                "PDSCH MCS decreased, which means the cell is using lower modulation. "
                "This usually happens due to weak radio quality, interference, or poor coverage. "
                "Check CQI/SINR, BLER, antenna tilt, azimuth, and interference sources."
            ),
            "action_priority": priority
        })

    # -------------------------------
    # Availability Issue
    # -------------------------------
    elif category == "Availability Issue":
        return pd.Series({
            "recommended_action_type": "Availability / Outage Investigation",
            "recommended_parameters_to_check": (
                "Availability, Cell Unavailable Time, System Unavailability, Manual Unavailability, S1 Failure"
            ),
            "detailed_recommended_action": (
                "Cell availability decreased. Check alarms, site outage, S1 failure, "
                "manual lock, system unavailability, transmission issue, and power issue. "
                "If availability is low, solve outage issue before RF optimization."
            ),
            "action_priority": priority
        })

    # -------------------------------
    # Cell Unavailability
    # -------------------------------
    elif category == "Cell Unavailability":
        return pd.Series({
            "recommended_action_type": "Cell Outage Investigation",
            "recommended_parameters_to_check": (
                "Cell Unavailable Time, Manual Unavailability, System Unavailability, S1 Failure, Alarms"
            ),
            "detailed_recommended_action": (
                "Cell unavailable time increased. Investigate whether the issue is caused by "
                "site outage, power issue, transmission failure, S1 failure, manual shutdown, or system fault."
            ),
            "action_priority": priority
        })

    # -------------------------------
    # Traffic Demand Drop
    # -------------------------------
    elif category == "Traffic Demand Drop":
        return pd.Series({
            "recommended_action_type": "Traffic Behavior Validation",
            "recommended_parameters_to_check": (
                "DL Active Users, Average UE Number, QCI-9 Traffic, Total DL Traffic, UL Traffic"
            ),
            "detailed_recommended_action": (
                "DL active users decreased, so the DL traffic drop may be caused by lower demand, "
                "not a real network degradation. Compare with UL traffic, QCI-9 traffic, "
                "active users, and nearby cells before applying RF actions."
            ),
            "action_priority": priority
        })

    # -------------------------------
    # Carrier Aggregation Issue
    # -------------------------------
    elif category == "Carrier Aggregation Issue":
        return pd.Series({
            "recommended_action_type": "Carrier Aggregation Optimization",
            "recommended_parameters_to_check": (
                "CA Activation, SCell Availability, CA Traffic Ratio, CA UE Number, CA Bands, SCell Addition Attempts"
            ),
            "detailed_recommended_action": (
                "CA traffic ratio decreased, which may reduce DL capacity and throughput. "
                "Check CA activation, SCell availability, CA UE count, CA bands, "
                "SCell addition/deactivation attempts, and CA feature configuration."
            ),
            "action_priority": priority
        })

    # -------------------------------
    # Default Bearer Traffic Drop
    # -------------------------------
    elif category == "Default Bearer Traffic Drop":
        return pd.Series({
            "recommended_action_type": "Service Traffic Investigation",
            "recommended_parameters_to_check": (
                "DL Traffic QCI-9, Active Users, Total DL Traffic, Packet Data Service, APN/Data Service"
            ),
            "detailed_recommended_action": (
                "QCI-9 DL traffic decreased, which may explain the total DL traffic drop. "
                "Check if this is caused by lower user demand, service issue, packet data problem, "
                "or abnormal traffic distribution."
            ),
            "action_priority": priority
        })

    # -------------------------------
    # Control Channel Congestion
    # -------------------------------
    elif category == "Control Channel Congestion":
        return pd.Series({
            "recommended_action_type": "PDCCH / CCE Optimization",
            "recommended_parameters_to_check": (
                "CCE Utilization, DL CCE Allocation Failure, PDCCH Capacity, Scheduler, Control Channel Configuration"
            ),
            "detailed_recommended_action": (
                "DL CCE allocation failure increased, which may affect scheduling and throughput. "
                "Check CCE utilization, PDCCH capacity, control channel configuration, and scheduler behavior."
            ),
            "action_priority": priority
        })

    # -------------------------------
    # Unknown Cause
    # -------------------------------
    else:
        return pd.Series({
            "recommended_action_type": "Manual Investigation",
            "recommended_parameters_to_check": (
                "Raw Counters, Alarms, Availability, Traffic Trend, Neighbor Cells, Recent Changes"
            ),
            "detailed_recommended_action": (
                "DL traffic degradation was detected, but no strong related cause was found from the selected counters. "
                "Check raw counters, alarms, recent parameter changes, availability, and nearby cells manually."
            ),
            "action_priority": priority
        })
```

```python
# ============================================================
# ALGORITHM 11 - APPLY RECOMMENDATION ENGINE
# ============================================================

recommendation_results = dl_degraded_cells_with_causes.apply(
    recommend_action_based_on_cause,
    axis=1
)

dl_degraded_cells_with_causes = pd.concat(
    [dl_degraded_cells_with_causes, recommendation_results],
    axis=1
)

print("Recommendation engine completed")
```

```python
# ============================================================
# ALGORITHM 12 - SELECT FINAL OUTPUT COLUMNS
# ============================================================

final_output_cols = [
    "eNodeB Name",
    "Cell Name",
    "LocalCell Id",

    "recent_avg_dl_traffic",
    "baseline_avg_dl_traffic",
    "recent_total_dl_traffic",
    "baseline_total_dl_traffic",
    "recent_days_count",
    "baseline_days_count",
    "dl_traffic_degradation_ratio_%",
    "dl_traffic_status",
    "degraded_kpi",

    "main_cause_counter_or_kpi",
    "main_cause_recent_value",
    "main_cause_baseline_value",
    "main_cause_change_%",
    "main_root_cause_category",
    "main_degradation_reason",
    "main_recommended_parameter_or_action",

    "number_of_detected_causes",
    "multi_cause_flag",
    "all_detected_causes",
    "all_cause_categories",
    "all_recommended_actions",

    "recommended_action_type",
    "recommended_parameters_to_check",
    "detailed_recommended_action",
    "action_priority"
]

final_dl_output = dl_degraded_cells_with_causes[final_output_cols].copy()

```

```python
# ============================================================
# ALGORITHM 13 - SAVE FINAL OUTPUT
# ============================================================


output_file = "dl_traffic_degraded_cells_with_causes_and_actions.csv"

final_dl_output.to_csv(output_file, index=False)

print("Final file saved:", output_file)
print("Number of degraded cells:", final_dl_output.shape[0])
final_dl_output.head()
```
